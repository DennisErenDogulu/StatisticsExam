### **Exercise 2.11: Computer chip control**

Vi definerer en ny random variable U for the time saving, som er:

$$U = 1* - 1*X_{new}$$

Vi bruger 2.56 til at finde mean og variance for U:




$$E(U) = E(X_{old}- X_{new}) = E(X_{old}) - E(X_{new}) = 65-54 = 11$$

Variansen er:

$$V(U) = V(X_{old}- X_{new}) = 1^2\cdot Var(X_{old}) + (-1)^2\cdot V(X_{new})=$$
$$ = V(X_{old}) + V(X_{new}) = 4^2 + 3^2 = 16 + 9 = 25$$


Dvs. $$U \sim N(11, 5^2)$$

```python
print("P(U < 10) =", stats.norm.cdf(x=10,loc=11,scale=np.sqrt(25)))
```

b) What is the mean (μ) and standard deviation (σ) for the total time use for checking 100 chips on the new machine is?


Vi laver en ny random variabel, T som er:

$$T = \sum_{i=1}^{100}X_i$$
hvor $X_i$ er de samme og $X_i \sim N(54, 3^2)$


$$
\begin{aligned}
\mu=\mathrm{E}(T)=\mathrm{E}\left(\sum_{i=1}^{100} X_i\right) & =\mathrm{E}\left(X_1+X_2+\cdots+X_{100}\right) \\
& =\mathrm{E}\left(X_1\right)+\mathrm{E}\left(X_2\right)+\cdots+\mathrm{E}\left(X_{100}\right) \\
& =\sum_{i=1}^{100} E\left(X_i\right)=\sum_{i=1}^{100} 54=100 \cdot 54=5400
\end{aligned}
$$

and

$$
\begin{aligned}
\sigma^2=\mathrm{V}(T)=\mathrm{V}\left(\sum_{i=1}^{100} X_i\right) & =\mathrm{V}\left(X_1+X_2+\cdots+X_{100}\right) \\
& =\mathrm{V}\left(X_1\right)+\mathrm{V}\left(X_2\right)+\cdots+\mathrm{V}\left(X_{100}\right) \\
& =\sum_{i=1}^{100} \mathrm{~V}\left(X_i\right)=\sum_{i=1}^{100} 9=100 \cdot 9
\end{aligned}
$$


So the answer is: $\mu=100 \cdot 54=5400 \mathrm{~ms}$ and $\sigma= \sqrt{900}=30 \mathrm{~ms}$.
